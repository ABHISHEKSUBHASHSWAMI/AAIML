```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import re

# Load the dataset
# Assuming the dataset is in a CSV file and contains columns "text" and "label"
dataset_path = 'email_dataset.csv'
data = pd.read_csv(dataset_path)

# Preprocess the data
# Define a function to clean the email text
def clean_email(text):
    # Convert the email to lowercase
    text = text.lower()
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    # Remove email addresses
    text = re.sub(r'\S+@\S+', '', text)
    # Remove special characters and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Remove extra spaces
    text = ' '.join(text.split())
    return text

# Apply the cleaning function to the "text" column
data['text'] = data['text'].apply(clean_email)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data['text'], data['label'], test_size=0.2, random_state=42)

# Vectorize the email text
# Using TF-IDF Vectorizer
vectorizer = TfidfVectorizer(max_features=10000)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Train a Random Forest classifier
classifier = RandomForestClassifier(n_estimators=100, random_state=42)
classifier.fit(X_train_vec, y_train)

# Make predictions on the test set
y_pred = classifier.predict(X_test_vec)

# Evaluate the model
print("\nModel Evaluation:")
print(classification_report(y_test, y_pred))
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")

# Save the trained model and vectorizer for future use
import joblib
model_path = 'email_spam_classifier.joblib'
vectorizer_path = 'email_vectorizer.joblib'
joblib.dump(classifier, model_path)
joblib.dump(vectorizer, vectorizer_path)
print(f"\nTrained model saved as {model_path}")
print(f"Vectorizer saved as {vectorizer_path}")
```

```python

# Function to classify new emails
def classify_new_email(email_text):
    # Load the trained model and vectorizer
    classifier = joblib.load(model_path)
    vectorizer = joblib.load(vectorizer_path)
    
    # Preprocess and clean the email text
    email_text_cleaned = clean_email(email_text)
    
    # Vectorize the email text
    email_text_vec = vectorizer.transform([email_text_cleaned])
    
    # Make prediction
    prediction = classifier.predict(email_text_vec)
    
    return prediction[0]

# Example usage: classify a new email
new_email_text = """
Hi John, 
I hope you are doing well!
"""
prediction = classify_new_email(new_email_text)
print(f"\nPrediction for the new email: {prediction}")
```
